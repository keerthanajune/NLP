## Natural Language Processing Lab

### Lab2. Computing Bigram Frequencies

#### EXCERCISE-1: Process simple bigram data file

#### STEP 1: OPEN the file, count_2w.txt

```python
f = open('count_2w.txt','r',encoding='utf-8')
data = f.readlines()
data
```

#### STEP 2: build goog2w_list

```python
goog2w_list = []
for m in data:
    (w1, w2, count) = m.split()
    count = int(count)
    goog2w_list.append(((w1, w2), count))
goog2w_list
```

```python
goog2w_list[0]
```

#### STEP 3: build goog2w_fd

```python
data
```

```python
import nltk
goog2w_fd = nltk.FreqDist()
for m in data:
    w1, w2, count = m.split()
    goog2w_fd[(w1,w2)]=count
goog2w_fd[('of','the')]
```

```python
import nltk 
goog2w_fd=nltk.FreqDist()
for m in data:
    w1, w2, count = m.split()
    goog2w_fd[(w1, w2)]=count
goog2w_fd[('so','beautiful')]
```

```python
goog2w_fd.most_common(10)
```

```python
import pickle

with open('goog2w_list.pkl','wb') as handle:
    pickle.dump(goog2w_list,handle,protocol=pickle.HIGHEST_PROTOCOL)
```
